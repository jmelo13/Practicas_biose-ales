```python
#a)

import numpy as np
import matplotlib.pyplot as plt

# Creaci√≥n de vectores usando NumPy
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])

print("Vector a:", a)
print("Vector b:", b)

#b)
producto_punto = np.dot(a,b);
print("Producto punto =",producto_punto)

#c)
multiplicacion_punto_punto = a*b;
print("Multiplicacion punto a punto =",multiplicacion_punto_punto)

#d)
A = np.array([[2,-1,-3],[4,1.5,-2.5],[7.3,-0.9,0.2]]);
print("Matriz A =",A)

#e)

A_transpuesta = A.T;
print("A transpuesta =",A_transpuesta)

#f)

#ones
# La funcion ones(np.ones) posibilita crear una matriz de n dimensiones, en la cual todos sus elementos son unos

B = np.ones((4,4));
print( "Matriz B =",B) 

#round
# La funcione round(np.round) redondea los elementos de una matriz existente, la cual posea elementos de valores que no sean exactos
C = np.array([2.456,1.928,7.564]);
print("Matriz C =",C)

Matriz_redondeada = np.round(C,1);
print("Matriz C redondeada =",Matriz_redondeada)

#ceil
# La funcion ceil(np.ceil) redoendea hacia arriba cada elemento del arreglo

Redondeo_hacia_arriba = np.ceil(C);
print("Matriz C redondeada hacia arriba =",Redondeo_hacia_arriba)

#floor
# La funcion florr(np.floor) redondea hacia abajo cada elemento del arreglo

Redondeo_hacia_abajo = np.floor(C);
print("Matriz C redondeada hacia abajo =",Redondeo_hacia_abajo)

#g)

valor = A[0,2];
print("Primera fila y tercera columna de la matriz A =",valor)

#h)

fila2 = A[1,:];
print("Segunda fila de la matriz A =",fila2)

#i)

Dimension = A.shape;
print("Tama√±o de la matriz A =",Dimension)

#j)

n = np.arange(0, 101);
y1 = np.sin(np.pi * 0.12 * n);

print("y1[n] =",y1[:10])

#k)

y2 = np.cos(np.pi * 0.03 * n);

print("y2[n] =",y2[:10])

#l)

#Tercera se√±al

S = y1 + y2;
print("S =",S)

#Cuarta se√±al

t = y1*y2;
print("t =",t)

#m)

plt.plot(n, y1, label="y[n] = sin(œÄ * 0.12 * n)", color='blue');
plt.plot(n, y2, label="y2[n] = cos(2œÄ * 0.03 * n)", color='orange');

# Asignar t√≠tulo a los ejes
plt.xlabel("n(√çndice)");
plt.ylabel("Amplitud");
# Asignar t√≠tulo a la gr√°fica
plt.title("Gr√°fica de y1 y y2");
# Mostrar leyenda para diferenciar las se√±ales
plt.legend();
# Mostrar la gr√°fica
plt.grid(True);
plt.show();

#n)

plt.plot(n, S, label="s[n] = y[n] + y2[n]", color='green')
plt.plot(n, t, label="t[n] = y[n] * y2[n]", color='red')
plt.xlabel("n (√çndice de Muestra)")
plt.ylabel("Amplitud")
plt.title("Gr√°fica de S y t")
plt.legend()

# Mostrar la gr√°fica
plt.grid(True)
plt.show()
```
